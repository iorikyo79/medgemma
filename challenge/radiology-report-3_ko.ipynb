```python
def generate_impression(findings_text):
    prompt = f"""### Instruction:
You are a board-certified radiologist. 
Given the "Findings" section of a chest X-ray report, write a concise, professional "Impression" section.
The Impression should:
- List the most important diagnoses first
- Use standardized terminology (e.g., "No acute cardiopulmonary abnormality")
- Avoid repeating normal findings unless clinically significant
- Be suitable for a referring physician


### Findings:
{findings_text}

### Impression:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        temperature=0.1,
        do_sample=False,
    )
    response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
    return response.strip()

# 첫 번째 보고서에 대한 테스트
test_findings = reports[0]["findings"]
generated = generate_impression(test_findings)
true_impression = reports[0]["impression"]

print(" Generated Impression:\n", generated)
print("\n Ground Truth Impression:\n", true_impression)
```

```python
val_dataset = load_dataset("rajpurkarlab/ReXGradient-160K", split="validation")
val_sample = val_dataset.shuffle(seed=42).select(range(200))

val_reports = []
for item in val_sample:
    val_reports.append({
        "findings": item.get("Findings", ""),
        "impression": item.get("Impression", "")
    })
```

```python
!pip install -q rouge-score bert-score

from rouge_score import rouge_scorer
from bert_score import score as bert_score

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# def evaluate_predictions(predictions, references):
#     rouge_scores = []
#     for pred, ref in zip(predictions, references):
#         scores = scorer.score(ref, pred)
#         rouge_scores.append(scores)
#     # 평균
#     avg_rouge = {}
#     for metric in ['rouge1', 'rouge2', 'rougeL']:
#         avg_rouge[metric] = sum([s[metric].fmeasure for s in rouge_scores]) / len(rouge_scores)
    
#     # BERTScore
#     P, R, F1 = bert_score(predictions, references, lang="en", verbose=True)
#     avg_bertscore = F1.mean().item()
    
#     return avg_rouge, avg_bertscore
def evaluate_predictions(predictions, references):
    rouge_scores = []
    for pred, ref in zip(predictions, references):
        scores = scorer.score(ref, pred)
        rouge_scores.append(scores)
    # 평균
    avg_rouge = {}
    for metric in ['rouge1', 'rouge2', 'rougeL']:
        avg_rouge[metric] = sum([s[metric].fmeasure for s in rouge_scores]) / len(rouge_scores)
    
    # CPU에서 BERTScore 실행 (속도는 느리지만 OOM 방지)
    P, R, F1 = bert_score(predictions, references, lang="en", verbose=True, device="cuda")
    avg_bertscore = F1.mean().item()
    
    return avg_rouge, avg_bertscore
    

# 검증 샘플에서 실행 (시간이 걸릴 수 있으므로 먼저 하위 집합에서 실행할 수 있습니다)
preds = []
refs = []
for i, item in enumerate(val_reports[:50]):  # start small
    print(f"Generating {i+1}/50")
    pred = generate_impression(item["findings"])
    preds.append(pred)
    refs.append(item["impression"])

rouge, bert = evaluate_predictions(preds, refs)
print("ROUGE:", rouge)
print("BERTScore F1:", bert)
```

```python
def generate_impression(findings_text):
    messages = [
        {
            "role": "user",
            "content": f"""You are a board-certified radiologist.
Given the "Findings" section of a chest X-ray report, write a concise, professional "Impression" section.
The Impression should:
- List the most important diagnoses first
- Use standardized terminology
- Be suitable for a referring physician

Findings:
{findings_text}"""
        }
    ]

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=128,
        temperature=0.1,
        do_sample=False,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id or tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
    return response.strip()
```

```python
import pandas as pd
import numpy as np

# CSV 파일 불러오기 (아직 불러오지 않은 경우)
csv_path = "/kaggle/input/datasets/hamidmomand/radiological-report/Cleaned_Diagnostic_Data.csv"
df = pd.read_csv(csv_path)

# 임의로 50개의 보고서 샘플링, 재현성을 위해 시드 설정
sample_size = 50
df_sample = df.sample(n=sample_size, random_state=42).reset_index(drop=True)

print(f" Sampled {len(df_sample)} reports")
```

```python
from tqdm import tqdm

# 결과를 저장할 리스트
predictions = []
references = []

for idx, row in tqdm(df_sample.iterrows(), total=len(df_sample), desc="Generating impressions"):
    findings = row["Details"]
    true_impression = row["Impression"]

    # 비어 있는 경우 건너뛰기 (혹시 모를 상황 대비)
    if pd.isna(findings) or pd.isna(true_impression):
        continue

    generated = generate_impression(findings)
    predictions.append(generated)
    references.append(true_impression)

print(f" Generated {len(predictions)} impressions")
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

rouge1_scores = []
rouge2_scores = []
rougeL_scores = []

for pred, ref in zip(predictions, references):
    scores = scorer.score(ref, pred)
    rouge1_scores.append(scores['rouge1'].fmeasure)
    rouge2_scores.append(scores['rouge2'].fmeasure)
    rougeL_scores.append(scores['rougeL'].fmeasure)

print("ROUGE Scores (average over {} reports):".format(len(predictions)))
print(f"ROUGE-1: {np.mean(rouge1_scores):.3f} (±{np.std(rouge1_scores):.3f})")
print(f"ROUGE-2: {np.mean(rouge2_scores):.3f} (±{np.std(rouge2_scores):.3f})")
print(f"ROUGE-L: {np.mean(rougeL_scores):.3f} (±{np.std(rougeL_scores):.3f})")
```

```python
# ... (기존 생성 루프) ...

# 모든 예측을 생성한 후, BERTScore를 위해 GPU를 비우고자 MedGemma를 CPU로 이동
model.to("cpu")
torch.cuda.empty_cache()

# 이제 평가 실행 (BERTScore는 자동으로 GPU를 사용합니다)
rouge, bert = evaluate_predictions(preds, refs)

# 나중에 필요한 경우 MedGemma를 다시 GPU로 이동
model.to("cuda")

print("ROUGE:", rouge)
print("BERTScore F1:", bert)
```
